## Libraries

```python
import rawpy as rw
import numpy as  np
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
```

## Reading the raw image

```python
with rw.imread('Images/scene_raw.CR2.dng') as raw:
    raw_image = raw.raw_image.copy()
    raw_colors = raw.raw_colors.copy()
    color_desc = raw.color_desc
    white_level = raw.camera_white_level_per_channel
    black_level = raw.black_level_per_channel
    size = raw.sizes
    image = raw.postprocess()
```

## Extract the channels from Bayes matrix

```python
color_desc
```

```python
white_level
```

```python
black_level
```

```python
def extract_channels(raw_image, raw_colors):
    
    n_channels = np.unique(raw_colors)
    
    bayes = np.zeros_like(raw_image)
    
    channel_dimension = np.zeros_like(n_channels)
    
    image_channels = bayes[:,:,np.newaxis] + channel_dimension[np.newaxis,np.newaxis,:]
    
    for channel in n_channels:
        mask = raw_colors != channel
        image_channels[:,:,channel] = raw_image.copy()
        image_channels[:,:,channel][mask] = 0
        
    return image_channels
```

```python
image_channels = extract_channels(raw_image, raw_colors)
```

## Linearization

```python
minimum = 255
maximum = 4000
```

```python
x = np.linspace(minimum+1, maximum, num = 1000)
```

```python
f = lambda x: (np.power(2,8*(x-minimum)/(maximum-minimum))).astype(int) - 1
```

```python
y = f(x)
```

```python
plt.plot(x,y);
```
